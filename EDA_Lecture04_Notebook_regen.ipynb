## 5. Structured vs Unstructured Data

- **Structured**: tables, rows, columns (like this Titanic dataset)
- **Unstructured**: text, images, audio, logs, etc.

We often turn unstructured data into structured features.

### 5.1 Simple text feature extraction

````python
texts = [
    "This Wednesday morn, are you early to rise? Then look East.",
    "The Crescent Moon joins Venus & Saturn.",
    "Afloat in the dawn skies."
]

df_text = pd.DataFrame({"text": texts})
df_text['length'] = df_text['text'].str.len()
df_text['word_count'] = df_text['text'].str.split().str.len()
df_text['has_question_mark'] = df_text['text'].str.contains('\?')
df_text
````

## 6. Univariate Data Analysis

Look at **one variable at a time**.

We will check distributions of age, fare, class, and sex.

````python
plt.figure()
titanic['age'].dropna().hist(bins=30)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Passenger Age')
plt.show()
````

````python
plt.figure()
titanic['fare'].dropna().hist(bins=30)
plt.xlabel('Fare')
plt.ylabel('Count')
plt.title('Distribution of Passenger Fare')
plt.show()
````

````python
plt.figure()
titanic['class'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('Class')
plt.ylabel('Number of Passengers')
plt.title('Passenger Count by Class')
plt.show()
````

````python
plt.figure()
titanic['sex'].value_counts().plot(kind='bar')
plt.xlabel('Sex')
plt.ylabel('Number of Passengers')
plt.title('Passenger Count by Sex')
plt.show()
````

## 7. Bivariate Data Analysis

Relationship between **two variables**.

Examples:
- Survival vs sex
- Age vs fare

````python
pd.crosstab(titanic['sex'], titanic['survived'], normalize='index')
````

````python
ct = pd.crosstab(titanic['sex'], titanic['survived'])
ct.plot(kind='bar', stacked=True)
plt.xlabel('Sex')
plt.ylabel('Number of Passengers')
plt.title('Survival by Sex')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
````

````python
plt.figure()
plt.scatter(titanic['age'], titanic['fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
plt.show()
````

## 8. Trivariate / Multivariate (3D+) Data

Add a third variable using color, size, or facets.

Here: age vs fare, colored by survival.

````python
fig, ax = plt.subplots()
survived_mask = titanic['survived'] == 1
ax.scatter(titanic.loc[~survived_mask, 'age'], titanic.loc[~survived_mask, 'fare'], label='Not survived', alpha=0.7)
ax.scatter(titanic.loc[survived_mask, 'age'], titanic.loc[survived_mask, 'fare'], label='Survived', alpha=0.7)
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
ax.set_title('Age vs Fare (colored by survival)')
ax.legend()
plt.show()
````
