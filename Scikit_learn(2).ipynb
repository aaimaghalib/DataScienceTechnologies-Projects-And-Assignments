# **Data Preprocessing in Scilit-learn**


Reality of real-world data

Real datasets usually contain:

Missing values

Different scales (age vs salary)

Categorical/text data

Mixed column types

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "age": [25, 30, np.nan, 45, 50],
    "salary": [50000, 60000, 55000, np.nan, 70000],
    "gender": ["Male", "Female", "Female", np.nan, "Male"],
    "city": ["NY", "LA", "NY", "SF", np.nan],
    "purchased": [0, 1, 1, 0, 1]
})
df
```

# **STEP 1 — Separate X and y**


Scikit-learn never mixes features and target.

Code

```python
X=df.drop("purchased",axis=1)
X
```

```python
Y=df["purchased"]
Y
```

# **STEP 2 — Identify Column Types**
Why?

Different preprocessing for:

Numeric columns

Categorical columns

```python
numeric_features=["age","salary"]
catagorical_features=["gender","city"]
```

# **Step 3 - Handles Missing Data**
sklearn.impute.SimpleImputer
What is imputation?

Imputation = replacing missing values with meaningful values.

Scikit-learn models:

❌ Cannot work with missing values
✅ Require imputed data

```python
from sklearn.impute import SimpleImputer
```

# **Numeric Imputer**
SimpleImputer(strategy="median")


What it does internally:

During fit():

Computes median of each numeric column

During transform():

Replaces NaN with the computed median



# **Categorical Imputer**

SimpleImputer(strategy="most_frequent")


What it does internally:

Finds the most common category in each column

Replaces missing values with it

## **Step -4 Feature Scaling**
What is scaling?

Scaling = making numeric features comparable.

Why scaling is important

Example:

Age:    20 – 80
Salary: 20,000 – 200,000


Without scaling:

Salary dominates distance-based models ❌

```python
from sklearn.preprocessing import StandardScaler
```

# **What it does internally**

During fit():

Computes mean and standard deviation for each column

During transform():

(x - mean) / std


Result:

Mean = 0

Standard deviation = 1

# **Step - 5 Encoding**


```python
from sklearn.preprocessing import OneHotEncoder
```

# **Step -6 Pipeline**

```python
from sklearn.pipeline import Pipeline
numeric_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])
```

Pipeline(steps=[ ... ])

Creates a pipeline object

steps must be a list of tuples
"imputer" → just a name (you choose it)

SimpleImputer(strategy="median"):

Learns median during fit()

Replaces missing values during transform()

📌 Output → numeric data with no missing values

```python
categorical_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])
```

# **Step-7 PIPELINE + COLUMNTRANSFORMER**
Why pipelines alone are not enough

Because:

Numeric columns need numeric pipeline

Categorical columns need categorical pipeline

That’s why we use ColumnTransformer.

```python
from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_pipeline, numeric_features),
        ("cat", categorical_pipeline, catagorical_features)
    ]
)
```

# **Apply Preprocessing**

# **1️⃣ fit() — “Learn from the data”**
What it means (very simple)

fit() means: “Look at the data and remember important things.”

It does not change the data.
It only learns rules.

Examples
Object	What fit() learns

Imputer	What value to replace missing data with

Scaler	Average and spread of numbers

Encoder	Which categories exist


```python
X_processed = preprocessor.fit(X)
X_processed
```

# **3️⃣ fit_transform() — “Learn + apply in one step”**

What it means

fit_transform() means: “Learn from the data and apply it immediately.”

This is just a shortcut.

```python
X_processed = preprocessor.fit_transform(X)
X_processed
```
