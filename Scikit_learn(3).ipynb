```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "age": [25, 30, np.nan, 45, 50, 35, 40, 22, 60, np.nan],
    "salary": [50000, 60000, 55000, np.nan, 70000, 65000, np.nan, 48000, 90000, 52000],
    "gender": ["Male", "Female", "Female", np.nan, "Male", "Female", "Male", "Female", "Male", "Male"],
    "city": ["NY", "LA", "NY", "SF", np.nan, "LA", "SF", "NY", "SF", "LA"],
    "purchased": [0, 1, 1, 0, 1, 1, 0, 0, 1, 0]
})
df
```

# **✅ STEP 2 — Split the dataset into X and y**

✅ Why do we split into X and y?

Because sklearn expects:

X → features (inputs)

y → target (output)

```python
X = df.drop("purchased", axis=1)
y = df["purchased"]
```

# **✅ STEP 3 — Train-Test Split**

✅ Why do we split train and test?

Because:

Training data is used to learn

Test data is used to check performance

If you train and test on same data:
✅ accuracy looks high
❌ but model is useless on new data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
)
```

# **✅ STEP 4 — Preprocessing**

```python
numeric_features = ["age", "salary"]
categorical_features = ["gender", "city"]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

numeric_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])
```

```python
from sklearn.preprocessing import OneHotEncoder

categorical_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])
```

```python
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_pipeline, numeric_features),
        ("cat", categorical_pipeline, categorical_features)
    ]
)
```

```python
X_train_processed = preprocessor.fit_transform(X_train)
```

```python
X_test_processed = preprocessor.transform(X_test)
```

# **✅ STEP 5 — Train Model Separately**

Now we train the model on the processed training data.

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train_processed, y_train)
```

model.fit() learns relationship between processed features and labels

Now the model is trained

# **✅ STEP 6 — Predict Separately**

```python
y_pred = model.predict(X_test_processed)
```

# **✅ STEP 7 — Evaluation Metrics**

**1️⃣ Confusion Matrix (Foundation of all classification metrics)**

✅ Definition

A confusion matrix is a table that shows:

how many predictions were correct

how many were wrong

and what type of mistake happened

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

✅ Confusion Matrix Form

For binary classification (0/1), it looks like:

	Predicted 0	Predicted 1
Actual 0    	TN	FP

Actual 1    	FN	TP

✅ Meaning of each term

✅ TP (True Positive)
Model predicted 1, and it was actually 1

✅ TN (True Negative)
Model predicted 0, and it was actually 0

❌ FP (False Positive)
Model predicted 1, but it was actually 0
(also called Type 1 error)

❌ FN (False Negative)
Model predicted 0, but it was actually 1
(also called Type 2 error)

# **2️⃣ Accuracy Score**
✅ Definition

Accuracy means:

“Out of all predictions, how many were correct?”

✅ Formula

Accuracy= TP+TN/
TP+TN+FP+FN

	​


```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

# **3️⃣ Precision**
✅ Definition

Precision means:

“Out of all predictions that the model said 1, how many were actually 1?”

✅ Formula

Precision= TP/
TP+FP

	​


```python
from sklearn.metrics import precision_score

precision = precision_score(y_test, y_pred)
print("Precision:", precision)
```

**4️⃣ Recall (Sensitivity / True Positive Rate)**
✅ Definition

Recall means:

“Out of all real 1 cases, how many did the model correctly detect?”

✅ Formula

Recall=TP/
TP+FN

	​


```python
from sklearn.metrics import recall_score

recall = recall_score(y_test, y_pred)
print("Recall:", recall)
```

# **5️⃣ F1 Score**
✅ Definition

F1 is a balance between precision and recall.

Used when you care about both.

✅ Formula

F1=2⋅Precision⋅Recall/
Precision+Recall

	​


```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_pred)
print("F1 Score:", f1)
```
