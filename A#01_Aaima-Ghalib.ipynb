```python
import numpy as np
import pandas as pd
```

```python
students_list = [
    [1, 'Ali', 20, 'Lahore'],
    [2, 'Sara', 21, 'Karachi'],
    [3, 'Ahmed', 19, 'Islamabad'],
    [4, 'Ayesha', 22, 'Lahore'],
    [5, 'Bilal', 20, 'Karachi']
]
cols = ['student_id', 'name', 'age', 'city']
df_std = pd.DataFrame(students_list, columns=cols)
```

```python
scores = [
    {'student_id': 1, 'subject': 'Maths', 'marks': 85},
    {'student_id': 2, 'subject': 'Maths', 'marks': 90},
    {'student_id': 3, 'subject': 'Maths', 'marks': 75},
    {'student_id': 1, 'subject': 'Physics', 'marks': 80},
    {'student_id': 4, 'subject': 'Physics', 'marks': 88},
    {'student_id': 5, 'subject': 'Maths', 'marks': 70}
]
df_scores = pd.DataFrame(scores)
```

```python
df_new_entries = pd.DataFrame({
    'student_id': [6, 7],
    'name': ['Hina', 'Usman'],
    'age': [23, 21],
    'city': ['Faisalabad', 'Lahore']
})
```

```python
print("Students Data:")
display(df_std.head())
print("Scores Data:")
display(df_scores.head())
print("Extra Students:")
display(df_new_entries)
```

<h3> Merging (Inner, Left, Right) </h3>

```python
merged_inner = df_std.merge(df_scores, on='student_id', how='inner')
merged_left = df_std.merge(df_scores, on='student_id', how='left')
merged_right = df_std.merge(df_scores, on='student_id', how='right')

display(merged_inner.head())
display(merged_left.head())
display(merged_right.head())
```

<h3>Concatenation</h3>

```python
combined_students = pd.concat([df_std, df_new_entries], axis=0, ignore_index=True)
display(combined_students.tail())
```

<h3> Groupby Analysis </h3>

```python
student_avg = merged_inner.groupby('name')[['marks']].agg('mean')
city_avg = merged_inner.groupby('city')[['marks']].mean()

print(f"Mean Marks per Student:\n{student_avg}")
print("-" * 30)
print(f"Mean Marks per City:\n{city_avg}")
```

<h3> Bonus (Region Mapping) </h3>

```python
region_map = pd.DataFrame([
    ('Lahore', 'Punjab'),
    ('Karachi', 'Sindh'),
    ('Islamabad', 'Capital'),
    ('Faisalabad', 'Punjab')
], columns=['city', 'region'])
```

```python
final_dataset = merged_inner.merge(region_map, on='city', how='left')

display(final_dataset.sample(5))
```
