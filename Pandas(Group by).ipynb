**How Groupby works internally?**


# 🔹 Split-Apply-Combine Strategy

The **Split-Apply-Combine** paradigm describes **how pandas `groupby` works internally**.

---

## 1️⃣ **Split**

* The data is **split into groups** based on a key.
* Usually, you use `groupby()` for this.

Example: split students by `city`:

```python
group = a.groupby('city')
```

* `group` now has **5 groups**: Lahore, Karachi, Islamabad, Sahiwal, Narowal
* Each group contains all rows for that city.

---

## 2️⃣ **Apply**

* Apply a **function** to each group separately.
* Functions can be: sum, mean, max, custom lambda functions, etc.

Example: calculate **average marks per city**:

```python
group['marks'].mean()
```

* pandas calculates the **mean of marks** in each city group individually.

---

## 3️⃣ **Combine**

* The results from each group are **combined back into a single DataFrame or Series**.

Example:

```
city
Karachi      57.4
Lahore       56.3
Islamabad    56.6
Sahiwal      56.7
narowal      89.0
Name: marks, dtype: float64
```

* You now have a **summary Series/DataFrame** with one row per group.
