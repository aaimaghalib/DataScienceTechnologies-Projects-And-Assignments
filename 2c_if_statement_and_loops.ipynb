If statement is used to take decision based on some condition.


```
if cond :
  statments
```



````python
a = 100
b = 30
if b > a:
    print("b is greater than a")
elif a == b:
    print("a and b are equal")
else:
    print("a is greater than b")
````

Loops

**While loop**

````python
i = 1
while i < 11:
    print(i)
    i += 1
````

**For loop**

````python
courses = ['deep learning', 'data mining', 'data science']
for course in courses:
    print(course)
````

````python
for x in range(6):
    print(x)
````

````python
for x in range(2, 6):
    print(x)
````

````python
for x in range(8, 2, -2):
    print(x)
````
