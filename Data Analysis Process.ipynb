# **Data Gathering**

```python
#1. opening a local csv file
import pandas as pd
pd.read_csv('students2.csv')
```

```python
# 2. Sep Parameter
pd.read_csv('students2.csv',sep=' ')
```

```python
#3. Index_col Parameter (customize index)
pd.read_csv('students2.csv',index_col='name')
```

```python
#4. use_cols parameter (specify the column during import time)
pd.read_csv('students2.csv', usecols=['marks'])
```

```python
#mutiple columns
pd.read_csv('students2.csv', usecols=['name', 'marks'])
```

```python
#skip the first n rows
pd.read_csv("students2.csv", skiprows=5)
```

```python
#skip specific rows
pd.read_csv("students2.csv", skiprows=[1, 2, 5])
```

```python
#read only first n rows
pd.read_csv("students2.csv", nrows=10)
```

```python
#na vlaue parameter
pd.read_csv("students2.csv", na_values=['-','no','M'])
```

```python
#read excel file
pd.read_excel('students2.xlsx')
```
