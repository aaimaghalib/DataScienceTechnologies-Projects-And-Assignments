

# **Concantenation AND Joins**



````python
#virtually stack two dataframes
import pandas as pd
Jan=pd.DataFrame({
    'sales': [10,80,90,40],
    'Day' : ['MON','TUE','WED','THUR']
})
Feb=pd.DataFrame({
    'sales': [80,80,40,40],
    'Day' : ['MON','TUE','WED','THUR']
})
print(Jan)
print(Feb)
````

````python
pd.concat([Jan,Feb])
````

````python
#indexes should not repaet
pd.concat([Jan,Feb],ignore_index=True)
````

````python
#Multiindex
multi=pd.concat([Jan,Feb],keys=['Jan','Feb'])
print(multi)
````

````python
multi.loc['Jan','Day']
````

````python
#horizontally stack
pd.concat([Jan,Feb],axis=1)
````

# **Joins/Merging**
     Join database on the basis of same column.



### **Joins in pandas**

Joins are used to **combine two DataFrames** based on a **common column(s)** (like SQL joins).

#### **Types of joins**

1. **Inner Join (`how='inner'`)**

   * Returns only rows that have **matching keys in both DataFrames**.
   * Syntax:

     ```python
     pd.merge(df1, df2, on='key', how='inner')
     ```

2. **Left Join (`how='left'`)**

   * Returns **all rows from the left DataFrame**, and matching rows from the right DataFrame.
   * Missing values are filled with `NaN`.
   * Syntax:

     ```python
     pd.merge(df1, df2, on='key', how='left')
     ```

3. **Right Join (`how='right'`)**

   * Returns **all rows from the right DataFrame**, and matching rows from the left DataFrame.
   * Missing values are filled with `NaN`.
   * Syntax:

     ```python
     pd.merge(df1, df2, on='key', how='right')
     ```

4. **Outer Join (`how='outer'`)**

   * Returns **all rows from both DataFrames**, filling `NaN` for missing matches.
   * Syntax:

     ```python
     pd.merge(df1, df2, on='key', how='outer')
     ```



````python

df1 = pd.DataFrame({
    'student_id': [1, 2, 3, 4, 5],
    'name': ['Ali', 'Sara', 'Hassan', 'Fatima', 'Zain'],
    'city': ['Lahore', 'Karachi', 'Lahore', 'Islamabad', 'Sahiwal']
})

df2 = pd.DataFrame({
    'student_id': [3, 4, 5, 6, 7],
    'marks': [78, 88, 95, 67, 80],
    'result': ['Pass', 'Pass', 'Pass', 'Fail', 'Pass']
})

print(df1)
print(df2)
````

````python
#inner join
pd.merge(df1,df2,on='student_id',how='inner')
````

````python
#left join
pd.merge(df1,df2,on='student_id',how='left')
````

````python
#right join
pd.merge(df1,df2,on='student_id',how='right')
````

````python
#full
pd.merge(df1,df2,on='student_id',how='outer')
````

# **Filteration**

````python
# 1. Find all students who scored more tahn 80.
````

````python
a1=pd.merge(df1,df2,on='student_id',how="inner")
a1[a1['marks'] > 80]
````

````python

# 2. find the students who have'nt taken test.
a2=pd.merge(df1,df2,on='student_id',how="left")
a2[a2['marks'].isna()]
````

````python
# 3. Find students who have both personal info and marks recorded and passed the exam (result == "Pass") and scored more than 90.
a1[(a1['marks'] > 90) & (a1['result'] == 'Pass')]
````

````python
#4 .  Find students whose names contain the letter "a" and have marks recorded.
a1[a1['name'].str.contains('a')]
````
