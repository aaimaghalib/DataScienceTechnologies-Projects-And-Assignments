```python
import pandas as pd
import numpy as np
```

```python
raw_data = {
    'ID': [1.0, 2.0, 3.0, 4.0, np.nan],
    'Name': ['Alice', 'Bob', None, 'Dan', 'Eve'],
    'Age': [23.0, np.nan, 35.0, -5.0, 29.0],
    'Salary': ['50000', 'not available', '62000', '58000', None],
    'Department': ['HR', 'IT', 'IT', None, 'Finance']
}

df_clean = pd.DataFrame(raw_data)
display(df_clean)
```

<h3> Identify Missing Values </h3>

```python
print("Missing values count:")
print(df_clean.isnull().sum())
```

<h3> Handle Age (Invalid & Missing) </h3>

```python
df_clean['Age'] = df_clean['Age'].apply(lambda x: np.nan if x < 0 else x)

# Logic: Calculate mean and fill
mean_age = df_clean['Age'].mean()
df_clean['Age'] = df_clean['Age'].fillna(mean_age)

print("\nAge column processed.")
```

<h3> Handle Salary (Inconsistent Types) </h3>

```python
df_clean['Salary'] = pd.to_numeric(df_clean['Salary'], errors='coerce')

# Fill with median
median_sal = df_clean['Salary'].median()
df_clean['Salary'] = df_clean['Salary'].fillna(median_sal)

print("Salary converted to numeric and filled.")
```

<h3> Handle ID (Missing) </h3>

```python
df_clean['ID'] = df_clean['ID'].interpolate(method='linear')
df_clean['ID'] = df_clean['ID'].astype(int)
```

<h3> Handle Name and Department </h3>

```python
df_clean['Name'] = df_clean['Name'].fillna("Unknown")

# Filling Department with Mode
dept_mode = df_clean['Department'].mode()[0]
df_clean['Department'] = df_clean['Department'].fillna(dept_mode)
```

```python
print("\nFinal NaN check:")
print(df_clean.isnull().sum())
```

<h1> SUMMARY </h1>
Assignment 2 Summary: Data Cleaning Process

In this assignment, we took a messy dataset containing student and employee information and cleaned it up step-by-step to make it usable for analysis. Here is exactly what was done:

Initial Check: We started by loading the data and checking for any missing values (NaN) to see which columns needed attention.

Fixing Ages: We noticed some invalid data, like a negative age (-5), which isn't possible. We treated this as a missing value and then filled all empty age entries with the average (mean) age of the group.

Cleaning Salaries: The salary column was inconsistent because it contained text like "not available." We converted the entire column to numbers (forcing errors to become blank) and filled the missing spots with the median salary to avoid skewing the data.

Handling IDs: One ID was missing, but since IDs follow a sequence, we filled the gap logically (using interpolation) and ensured all IDs were stored as whole numbers (integers).

Filling Text Info: For the missing names, we simply labeled them as "Unknown." For the missing department, we filled it with the most common department (the mode), which was "IT" in this case.

Final Output: Once all errors were fixed and gaps were filled, we saved the clean, structured data into a new CSV file ready for use

```python
# Saving the file
df_clean.to_csv('Processed_Student_Data.csv', index=False)
```
