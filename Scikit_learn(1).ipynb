# **1. Importing a Dataset (Built-in Dataset)**


Scikit-learn provides ready-made datasets for learning.

```python
from sklearn.datasets import load_diabetes
data=load_diabetes()
```

### **2. Understanding the Dataset Structure**


A sklearn dataset usually contains:

data → feature values

target → labels

feature_names → column names

target_names → class names

```python
print(data.keys())
```

# **3. Extracting Features (X) and Target (y)**


Machine learning needs:

X → features

y → target

```python
X = data.data
y = data.target
```

# **4. Checking Shape of X and y**


Shape confirms correct data format.

```python
print(X.shape)
print(y.shape)
# X.shape → (number of rows, number of features)

# y.shape → (number of rows,)
```

# **5. Converting Data to Pandas**


Pandas makes data easier to read and analyze.

```python
import pandas as pd
df1=pd.DataFrame(X,columns=data.feature_names)
df1
```

```python
df2=pd.Series(y,name='Target_column')
df2
```

# **6. What is Train-Test Splitting**


We divide data into:

Training set → model learns

Testing set → model evaluation

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(
    df1,
    df2,
    test_size=0.2,
)
```

```python
print(X_train.shape)
print(X_test.shape)
```

train_test_split() → splits data

test_size=0.2 → 20% test data
